# Get data model ready

```python
# we have to drop ID, Zipcode from the data, because it barely provides insight
# and we also have to drop negative value of experience because we do not know how it comes.
data = df.drop(['ID','ZIP Code','len_ZIP'],axis=1)
```

```python
data = data.query('Experience >= 0')
```

```python
data.head()
```

# Split the data into training and test set in the ratio of 70:30 respectively

```python
X = data[['Age', 'Experience', 'Income', 'Family', 'CCAvg', 'Education',
       'Mortgage', 'Securities Account', 'CD Account',
       'Online', 'CreditCard']]
y = data['Personal Loan']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=1111,stratify=y)
```

# Modeling

## Use the Logistic Regression model to predict

```python
from sklearn.preprocessing import MinMaxScaler, OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import FunctionTransformer
from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer,precision_score
import warnings
warnings.filterwarnings('ignore')
```

```python
# Create log transformation function
def log_transform(x):
    return np.log(x + 1)
```

```python
# Create 3 transformer pipelines, one for unskewed numerical data, one for categorical data, one for skewed numerical data
numeric_transformer = Pipeline(steps=[
       ('scaler', MinMaxScaler()),
    ('transformer', FunctionTransformer(log_transform))
])
categorical_transformer = Pipeline(steps=[
       ('encoder', OrdinalEncoder())
])
skewed_transformer = Pipeline(steps=[
    ('transformer', FunctionTransformer(log_transform)),
    ('scaler', MinMaxScaler())
])
```

```python
numeric_features = ['Age', 'Experience', 'Family']
categorical_features = ['Education', 'Securities Account', 'CD Account','Online', 'CreditCard']
skewed_features = ['Income', 'CCAvg','Mortgage']
preprocessor = ColumnTransformer(
   transformers=[
    ('numeric', numeric_transformer, numeric_features)
   ,('categorical', categorical_transformer, categorical_features)
   ,('skewd_num', skewed_transformer, skewed_features)
]) 
```

```python
# Create model pipeline using LogisticRegression
pipeline = Pipeline(steps = [
               ('preprocessor', preprocessor)
              ,('regressor',LogisticRegression(solver='lbfgs', max_iter=1000))
           ])
```

```python
# List all the metrices
scoring = {"AUC": "roc_auc", "Accuracy": "accuracy" ,"recall":"recall", "precision":"precision","f1":"f1"}
```

```python
parameters = {
    "regressor__C": np.logspace(-4, 4, 4)
  }
```

```python
# Using GridSearch Cross Validation in order to find the best model.
logreg_cv=GridSearchCV(pipeline,parameters,scoring=scoring,cv=10,refit="AUC",return_train_score=True,)
```

```python
# Model Training
logreg_cv.fit(X_train,y_train)
```

```python
# Predict Value
y_pred = logreg_cv.predict(X_test)
```

```python
# Best C value for the LogisticRegression model
logreg_cv.best_params_
```

```python
# Best accuracy score
logreg_cv.score(X_test, y_test)
```

```python
results =logreg_cv.cv_results_
```

```python
print('Best score for every Cross-validation fold with test dataset')
print('AUC score is {}'.format(max(results['mean_test_AUC'])))
print('Accuracy score is {}'.format(max(results['mean_test_Accuracy'])))
print('Recall score is {}'.format(max(results['mean_test_recall'])))
print('Precision score is {}'.format(max(results['mean_test_precision'])))
print('F1 score is {}'.format(max(results['mean_test_f1'])))
```

```python
print('Average score for every Cross-validation fold with test dataset')
print('AUC score is {}'.format(np.mean(results['mean_test_AUC'])))
print('Accuracy score is {}'.format(np.mean(results['mean_test_Accuracy'])))
print('Recall score is {}'.format(np.mean(results['mean_test_recall'])))
print('Precision score is {}'.format(np.mean(results['mean_test_precision'])))
print('F1 score is {}'.format(np.mean(results['mean_test_f1'])))
```

```python
x_axis_val = [0.0001,0.046415888336127774,21.54434690031882,10000.0]

plt.figure(figsize=(13, 13))
plt.title("GridSearchCV evaluating using multiple scorers simultaneously", fontsize=16)

plt.xlabel("C_Values")
plt.ylabel("Score")

ax = plt.gca()
ax.set_xlim(0, 402)
ax.set_ylim(0.5, 1)

# Get the regular numpy array from the MaskedArray
X_axis = np.array(x_axis_val, dtype=float)

for scorer, color in zip(sorted(scoring), ["g", "k", "b","r","m"]):
    for sample, style in (("train", "--"), ("test", "-")):
        sample_score_mean = results["mean_%s_%s" % (sample, scorer)]
        sample_score_std = results["std_%s_%s" % (sample, scorer)]
        ax.fill_between(
            X_axis,
            sample_score_mean - sample_score_std,
            sample_score_mean + sample_score_std,
            alpha=0.1 if sample == "test" else 0,
            color=color,
        )
        ax.plot(
            X_axis,
            sample_score_mean,
            style,
            color=color,
            alpha=1 if sample == "test" else 0.7,
            label="%s (%s)" % (scorer, sample),
        )

    best_index = np.nonzero(results["rank_test_%s" % scorer] == 1)[0][0]
    best_score = results["mean_test_%s" % scorer][best_index]

    # Plot a dotted vertical line at the best score for that scorer marked by x
    ax.plot(
        [
            X_axis[best_index],
        ]
        * 2,
        [0, best_score],
        linestyle="-.",
        color=color,
        marker="x",
        markeredgewidth=3,
        ms=8,
    )

    # Annotate the best score for that scorer
    ax.annotate("%0.2f" % best_score, (X_axis[best_index], best_score + 0.005))

plt.legend(loc="best")
plt.grid(False)
plt.show()
```

# Give conclusion related to the Business understanding of your model?

According to coefficients of all the attributes (result_df), <br>
The top 3 most significant **numerical attributes** are **Income, CCAvg and Mortgage.** <br> 
Which means that **the customers annual income is the most significant numerical factor** that leads to high personal loan acceptance rate. <br>
At the same time, **average spending on credit cards per month and Value of house mortgage** have **high negative impacts** on persoanl loan acceptance rate.

The top 4 most significant **categorical attributes** are **CD Account, Education, Online and Family.**
Which means that **owning a certificate of deposit (CD) account with the bank** has **high positive impacts** on persoanl loan acceptance rate, followed by Education Level, Annual income, internet banking facility usage and Family size.
